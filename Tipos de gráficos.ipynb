# TIPOS DE GRÁFICOS EN PYTHON

```python
# Librerías básicas
import numpy as np
import pandas as pd

# Librerías para gráficos
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Librería para obtener datasets
from sklearn import datasets
```

```python
# Importar data local
data = pd.read_csv('countryGDP.csv')
```

```python
# Eliminando los missings
data1 = data['GDP per Capita'].dropna()
```

## 1- Histogramas

```python
plt.hist(data1, bins=50)
plt.show()
# bins para la cantidad de divisiones
```

```python
plt.hist(data1, bins=50, alpha=0.5)
plt.show()
# alpha para la opacidad
```

```python
plt.hist(data1, bins=50, alpha=0.5, color='green')
plt.title('PBI Nacional')
plt.xlabel('PBI en dólares')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.hist(data1, bins=50, alpha=0.5, color='green',
         orientation='horizontal')
plt.show()
```

```python
plt.hist(data1, bins=50, alpha=0.5, color='green',
         histtype='step')
plt.show()
```

## 2 - Análisis multivariado

```python
tips = sns.load_dataset('tips')
```

```python
g = sns.FacetGrid(tips, row='smoker', col='time')
g.map(plt.hist, 'total_bill')
```

```python
day = sns.FacetGrid(tips, row='day', col='time')
day.map(plt.hist, 'total_bill')
```

## 3 - Boxplots

```python
iris = sns.load_dataset('iris')
```

### Gráficos univariados

```python
iris.boxplot()
plt.show()
```

```python
iris.hist()
plt.show()
```

### Gráficos multivariados

```python
iris.boxplot(column='sepal_length', by='species')
plt.show()
```

```python
iris.boxplot(column='sepal_width', by='species')
plt.show()
```

```python
iris.boxplot(column='petal_length', by='species')
plt.show()
```

```python
iris.boxplot(column='petal_width', by='species')
plt.show()
```

```python
sns.boxplot(x='sex', y='total_bill', hue='time', data=tips)
plt.show()
```

## 4 - Scatter

```python
plt.scatter(tips.total_bill, tips.tip)
plt.show()
```

```python
plt.scatter(tips.total_bill, tips.tip, marker='x')
plt.show()
```

```python
plt.scatter(tips.total_bill, tips.tip, marker='x',
            alpha=0.5)
plt.show()
```

```python
plt.scatter(tips.total_bill, tips.tip, marker='x',
            alpha=0.5, s=100)
plt.show()
```

```python
plt.scatter(tips.total_bill, tips.tip, marker='x',
            alpha=0.5, s=100, color='green')
plt.show()
```

```python
# Importando libreria
from pandas.tools.plotting import scatter_matrix
scatter_matrix(iris)
plt.show()
```

```python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
g = sns.FacetGrid(iris, hue='species', height=5)
g.map(plt.scatter, 'sepal_length',
      'sepal_width').add_legend()
plt.show()
```

## 6 - Pie Chart

```python
# Composición
feliz['Region'].value_counts()
```

## 8 - Funciones de densidad

```python
sns.distplot(data1, hist=True, kde=True, color='blue', bins=50)
plt.show()
```

```python
sns.distplot(data1, hist=False, kde=True, color='blue', bins=50)
plt.show()
```

```python
sns.distplot(data1, hist=True, kde=False, color='blue', bins=50)
plt.show()
```

```python
sns.distplot(data1, hist=True, kde=True, color='blue', bins=50,
             hist_kws={'edgecolor':'black'},
             vertical=True)
plt.show()
```
